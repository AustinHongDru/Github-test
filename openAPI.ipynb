```python
def openurl(url) :
    
    headers = {'User-Agent' : 'Chrome/66.0.3359.181'}
    req = urllib.request.Request(url, headers= headers)
    data = urllib.request.urlopen(req)
        
    source = data.read()
    soup = BeautifulSoup(source, "html.parser")
    print(soup)
```
